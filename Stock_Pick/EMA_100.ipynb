```python
import pandas as pd
import yfinance as yf
import talib as ta
import numpy as np
```

```python
stock_names = pd.read_csv('All_Equity.csv')
names = stock_names['Symbol']
```

```python
Stock_picked = []
for symbol in names:
    Symbol = symbol+'.NS'

    try:
        stock_data = yf.download(Symbol, period='1y')
        Open = np.array(stock_data.loc[:, 'Open'].values)
        Close = np.array(stock_data.loc[:, 'Close'].values)
        High = np.array(stock_data.loc[:, 'High'].values)
        Low = np.array(stock_data.loc[:, 'Low'].values)
        
        ema_40 = ta.EMA(Close, timeperiod = 100)

        ##Logic
        if 50 <= Close[-1] < 600:
            if Low[-1] <= ema_40[-1] and High[-1] > ema_40[-1]:
                Stock_picked.append(symbol)
                print(f"Stock {symbol} has been added to the picked list....")
                
    except Exception as e:
        print(e)
```

```python
print(Stock_picked)
df = pd.DataFrame({"name": Stock_picked})
data_to = "EMA_40 Picked stocks.csv"
df.to_csv(data_to, index = False)
```
