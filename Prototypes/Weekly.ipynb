```python
import yfinance as yf
import pandas as pd
import warnings
import talib as ta

# Suppress FutureWarning related to pandas Series indexing
warnings.simplefilter(action='ignore', category=FutureWarning)


def get_stock_name(file_name):
    below_400 = []
    file = pd.read_csv(f"{file_name}.csv")
    for symbol in file.iloc[:, 0].tolist():
        Symbol = symbol + ".NS"
        stock_data = yf.download(Symbol, period="1y", interval="1wk")

        try:
            open = stock_data.loc[:, "Open"].tolist()[-1]
            low = stock_data.loc[:, "Low"].tolist()[-1]
            close = stock_data.loc[:, "Close"].tolist()[-1]
            sma = ta.SMA(stock_data.loc[:, "Close"].values, timeperiod=40)
        except Exception as e:
            print(e)

        if open >= 50 and open <= 400 and close > open:
            if low <= sma[-1] and close >= sma[-1]:
                below_400.append(symbol)
                print(symbol)

    for i in below_400:
        print(i)
    
get_stock_name("Input/All_Equity")
```
